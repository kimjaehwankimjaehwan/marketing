```python
import numpy as np
import matplotlib.pyplot as plt

# ÏÑ§Ï†ïÎêú ÌååÎùºÎØ∏ÌÑ∞Îì§
alpha = 0.88  # Ïù¥ÎìùÏóê ÎåÄÌïú Í∞êÏàòÏÑ±
beta = 0.88   # ÏÜêÏã§Ïóê ÎåÄÌïú Í∞êÏàòÏÑ±
lambda_ = 2.25  # ÏÜêÏã§ ÌöåÌîº Í≥ÑÏàò

# Í∞í ÏÉùÏÑ±
x = np.linspace(-1, 1, 400)  # -1ÏóêÏÑú 1ÍπåÏßÄÏùò x Í∞í ÏÉùÏÑ±
y_positive = x[x >= 0] ** alpha  # Ïù¥Îìù Î∂ÄÎ∂Ñ
y_negative = -lambda_ * (-x[x < 0]) ** beta  # ÏÜêÏã§ Î∂ÄÎ∂Ñ

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(7, 6))
plt.plot(x[x >= 0], y_positive, label="Ïù¥Îìù (Gain)", color="blue")
plt.plot(x[x < 0], y_negative, label="ÏÜêÏã§ (Loss)", color="red")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Ï†ÑÎßùÏù¥Î°†Ïùò Í∞ÄÏπò Ìï®Ïàò")
plt.xlabel("Í≤∞Í≥º (Outcome)")
plt.ylabel("Í∞ÄÏπò (Value)")
plt.legend()
plt.grid(True)
plt.show()
```

2. ÌôïÎ•† Í∞ÄÏ§ë Ìï®Ïàò(Probability Weighting Function)

```python
import numpy as np
import matplotlib.pyplot as plt

# ÏÑ§Ï†ïÎêú ÌååÎùºÎØ∏ÌÑ∞
gamma = 0.61  # ÌôïÎ•† ÏôúÍ≥° Ï†ïÎèÑ

# ÌôïÎ•† Í∞í ÏÉùÏÑ±
p = np.linspace(0, 1, 400)  # 0ÏóêÏÑú 1ÍπåÏßÄÏùò ÌôïÎ•† Í∞í ÏÉùÏÑ±
pi_p = p ** gamma / (p ** gamma + (1 - p) ** gamma) ** (1 / gamma)  # ÌôïÎ•† Í∞ÄÏ§ëÏπò Í≥ÑÏÇ∞

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(7, 6))
plt.plot(p, pi_p, label="ÌôïÎ•† Í∞ÄÏ§ëÏπò (Probability Weight)", color="green")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.plot(p, p, '--', label="Ïã§Ï†ú ÌôïÎ•† (Actual Probability)", color="black")
plt.title("Ï†ÑÎßùÏù¥Î°†Ïùò ÌôïÎ•† Í∞ÄÏ§ë Ìï®Ïàò")
plt.xlabel("Ïã§Ï†ú ÌôïÎ•† (Actual Probability)")
plt.ylabel("Í∞ÄÏ§ë ÌôïÎ•† (Weighted Probability)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ÏÑ§Ï†ïÎêú ÌååÎùºÎØ∏ÌÑ∞Îì§
lambda_ = 2.25  # ÏÜêÏã§ ÌöåÌîº Í≥ÑÏàò

# Í∞í ÏÉùÏÑ±
x = np.linspace(-0.99, 0.99, 400)  # Î°úÍ∑∏ Ìï®ÏàòÍ∞Ä Ï†ïÏùòÎêòÏßÄ ÏïäÎäî Î∂ÄÎ∂Ñ(-1 Ïù¥ÏÉÅ) ÌîºÌï®
y_positive_log = np.log(1 + x[x >= 0])  # Ïù¥Îìù Î∂ÄÎ∂Ñ
y_negative_log = -lambda_ * np.log(1 - x[x < 0])  # ÏÜêÏã§ Î∂ÄÎ∂Ñ

# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(7, 6))
plt.plot(x[x >= 0], y_positive_log, label="Ïù¥Îìù (Gain)", color="blue")
plt.plot(x[x < 0], y_negative_log, label="ÏÜêÏã§ (Loss)", color="red")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Î°úÍ∑∏ Ìï®ÏàòÎ•º Ï†ÅÏö©Ìïú Ï†ÑÎßùÏù¥Î°†Ïùò Í∞ÄÏπò Ìï®Ïàò")
plt.xlabel("Í≤∞Í≥º (Outcome)")
plt.ylabel("Í∞ÄÏπò (Value)")
plt.legend()
plt.grid(True)
plt.show()
```
