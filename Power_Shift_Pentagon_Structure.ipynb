```python
import matplotlib.pyplot as plt
import numpy as np

# 각 요소의 점수 예시 (0~10 사이)
labels = ['Consumer Behavior', 'Competitive Environment', 'Technological Change', 'Regulatory Environment', 'Distribution Channel']
values = [8, 6, 7, 5, 9]

# 오각형 차트 그리기 위한 설정
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
values += values[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# 차트 그리기
ax.fill(angles, values, color='skyblue', alpha=0.4)
ax.plot(angles, values, color='blue', linewidth=2)  # 외곽선
ax.set_yticks([2, 4, 6, 8, 10])  # 눈금
ax.set_yticklabels([])  # 눈금 라벨 제거
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

plt.title('Power Shift Pentagon Structure (PSPS) Visualization')
plt.show()
```
