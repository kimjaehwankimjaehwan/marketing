R의 fmsb 패키지를 사용한 오각형 차트(레이더 차트)

```r
# fmsb 패키지 설치
install.packages("fmsb")
library(fmsb)

# 데이터 준비: 최소값과 최대값을 포함한 데이터 프레임 생성
data <- data.frame(
  ConsumerBehavior = c(10, 0, 8),
  CompetitiveEnvironment = c(10, 0, 6),
  TechnologicalChange = c(10, 0, 7),
  RegulatoryEnvironment = c(10, 0, 5),
  DistributionChannel = c(10, 0, 9)
)

# 레이더 차트 그리기
radarchart(data,
           axistype = 1,
           pcol = "blue",
           pfcol = rgb(0.2, 0.5, 0.5, 0.5),
           plwd = 2,
           cglcol = "grey",
           cglty = 1,
           axislabcol = "grey",
           cglwd = 0.8,
           vlcex = 0.8
)

# 제목 추가
title("Power Shift Pentagon Structure (PSPS) Visualization")
```
