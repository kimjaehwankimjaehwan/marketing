```matlab
% 데이터 준비: 5개의 요소 점수
values = [8, 6, 7, 5, 9];

% 각도 계산 및 데이터 설정
angles = linspace(0, 2 * pi, 6); % 5개의 요소에 대해 각도 계산
values = [values values(1)]; % 오각형을 완성하기 위해 첫 번째 값 추가
angles(end) = angles(1); % 첫 번째 각도로 돌아가기 위해 추가

% 차트 그리기
figure;
polarplot(angles, values, '-o', 'LineWidth', 2);
ax = gca;
ax.ThetaTickLabel = {'Consumer Behavior', 'Competitive Environment', 'Technological Change', 'Regulatory Environment', 'Distribution Channel'};

% 레이더 차트 스타일 설정
rticks([2 4 6 8 10]) % 레이더 차트 눈금
title('Power Shift Pentagon Structure (PSPS) Visualization');
```
